```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import talib as ta
import os

# Ensure notebooks and processed directories exist
notebook_dir = os.path.join(os.getcwd(), 'notebooks')
processed_dir = os.path.join(os.getcwd(), 'data', 'processed')
os.makedirs(notebook_dir, exist_ok=True)
os.makedirs(processed_dir, exist_ok=True)

# Debug: Print current working directory and processed directory
print(f"Current working directory: {os.getcwd()}")
print(f"Processed directory: {processed_dir}")

# Stocks to analyze
tickers = ['AAPL','AMZN', 'GOOG', 'META','MSFT','NVDA', 'TSLA']

# Process each stock
for ticker in tickers:
    try:
        # Load stock data
        stock_data = pd.read_csv(f'../data/yfinance_data/{ticker}.csv')
        stock_data['Date'] = pd.to_datetime(stock_data['Date'])
        stock_data.set_index('Date', inplace=True)

        # Calculate indicators using TA-Lib
        stock_data['SMA_20'] = ta.SMA(stock_data['Close'], timeperiod=20)
        stock_data['RSI_14'] = ta.RSI(stock_data['Close'], timeperiod=14)
        stock_data['MACD'], stock_data['MACD_Signal'], stock_data['MACD_Hist'] = ta.MACD(
            stock_data['Close'], fastperiod=12, slowperiod=26, signalperiod=9
        )

        # Display results
        print(f"\n{ticker} Technical Indicators (Last 5 rows):")
        print(stock_data[['Close', 'SMA_20', 'RSI_14', 'MACD', 'MACD_Signal', 'MACD_Hist']].tail())

        # Visualization: Close Price and SMA
        plt.figure(figsize=(12, 4))
        plt.plot(stock_data['Close'], label='Close Price', color='#1f77b4')
        plt.plot(stock_data['SMA_20'], label='20-Day SMA', color='#ff7f0e')
        plt.title(f'{ticker} Close Price and 20-Day SMA')
        plt.xlabel('Date')
        plt.ylabel('Price (USD)')
        plt.legend()
        plt.tight_layout()
        plt.savefig(os.path.join(notebook_dir, f'{ticker}_price_sma.png'))
        plt.show()

        # Visualization: RSI
        plt.figure(figsize=(12, 4))
        plt.plot(stock_data['RSI_14'], label='RSI (14)', color='#2ca02c')
        plt.axhline(70, color='red', linestyle='--', alpha=0.5)
        plt.axhline(30, color='green', linestyle='--', alpha=0.5)
        plt.title(f'{ticker} RSI (14)')
        plt.xlabel('Date')
        plt.ylabel('RSI')
        plt.legend()
        plt.tight_layout()
        plt.savefig(os.path.join(notebook_dir, f'{ticker}_rsi.png'))
        plt.show()

        # Visualization: MACD
        plt.figure(figsize=(12, 4))
        plt.plot(stock_data['MACD'], label='MACD', color='#1f77b4')
        plt.plot(stock_data['MACD_Signal'], label='Signal Line', color='#ff7f0e')
        plt.bar(stock_data.index, stock_data['MACD_Hist'], label='Histogram', color='#2ca02c', alpha=0.5)
        plt.title(f'{ticker} MACD')
        plt.xlabel('Date')
        plt.ylabel('MACD')
        plt.legend()
        plt.tight_layout()
        plt.savefig(os.path.join(notebook_dir, f'{ticker}_macd.png'))
        plt.show()

        # Save processed data
        stock_data.to_csv(os.path.join(processed_dir, f'{ticker}_data.csv'))

    except FileNotFoundError:
        print(f"Error: {ticker}.csv not found in data/yfinance_data/")
    except Exception as e:
        print(f"Error processing {ticker}: {e}")
```
